## Сортировка подсчётом

```python
# Сложность у нас O(n+k)
def counting_sort(arr, k):
    tmp = [0] * (k+1)
    n = len(arr)
    for key, _ in arr:
        tmp[key] += 1
    for i in range(1, k+1):
        tmp[i] += tmp[i-1]
    res = [0] * n
    for j in range(n-1, -1, -1):
        key = arr[j][0]
        ridx = tmp[key]-1
        res[ridx] = arr[j]
        tmp[key] -= 1
    return res
```

```python
import random
import string

def generate_item(max_key_value=2**32-1):
    key = random.randint(0, max_key_value)
    value = random.choice(string.ascii_uppercase)
    return key, value
```

```python
N = 10
arr = [generate_item(10) for _ in range(N)]
k = max([elem[0] for elem in arr])
print(arr, k)
```

```python
sorted_arr = counting_sort(arr, k)
```

## Карманная сортировка

```python
import numpy
import random

def bucket_sort(arr):
    n = len(arr)
    buckets = []
    for i in range(n):
        buckets.append([])
    f = lambda num, length: int(num * length)
    for num in arr:
        idx = f(num, n)
        buckets[idx].append(num)
    for bucket in buckets:
        print(bucket)
        # Timsort = merge sort + insertion sort
        # Для маленьких данных и такое сгодится.
        # Лучше использовать чистую сортировку вставками.
        bucket.sort()
    res = []
    for bucket in buckets:
        for num in bucket:
            res.append(num)
    return res
```

```python
N = 1000
#arr = [random.uniform(0.0, 1.0) for _ in range(N)]
arr = numpy.random.uniform(0.0, 1.0, N)
#arr = [random.random() for _ in range(N)]
print(arr[:10])
```

```python
sorted_arr = bucket_sort(arr)
print(sorted_arr[:10])
```
